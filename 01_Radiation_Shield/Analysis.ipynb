```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Histograms parameters.
N_BINS = 100
MAX_ENERGY = 0.8 # MeV

filename = "build/RadShield_Data_h1_LeakageSpectrum.csv" 

# Read the CSV file, skipping Geant4 header lines.
df = pd.read_csv(filename, comment='#')

counts = df['entries']

# Create the energy axis, maps 0 to 800 keV across the number of bins in the file.
energy_kev = np.linspace(0, MAX_ENERGY * 1000, len(counts))

plt.figure(figsize=(10, 6))

# Plot.
plt.step(energy_kev, counts, where='mid', color='blue', label='Data')

total_particles = int(counts.sum())
plt.title(f"Radiation Leakage Spectrum\nTotal Particles Detected: {total_particles}", fontsize=14)
plt.xlabel("Energy [keV]", fontsize=12)
plt.ylabel("Counts per Bin", fontsize=12)

plt.xlim(0, 800)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()

# Data analysis output

peak_idx = counts.idxmax()
peak_energy = energy_kev[peak_idx]

print(f"--- Simulation Results ---")
print(f"Total particles detected: {total_particles}")
print(f"Peak value (max counts): {int(counts.max())}")
print(f"Peak energy position: {peak_energy:.2f} keV")
```
